```python
#import the necessary libraries
import matplotlib.pyplot as plt

#creating a figure
fig = plt.figure(figsize = (12, 5))
#show the plot
fig.show()
```

```python
#creating plot
#1. line plot

x = [1,3,5,4,7]
y = [2,5,7,8,10]


plt.plot(x,y)
plt.show()
```

```python
# costomizing plot
# it includes: titles, labes,font size

# creating plot

x = [1,2,3,4,5]
y = [2,5,7,8,10]

plt.title("line plot", fontsize=200)
plt.plot(x,y)
plt.show()
```

```python
#creating plot
#1. line plot

x = [1,2,3,4,5]
y1 = [2,5,7,8,10]
y2 = [1,3,5,7,9]

plt.plot(x, y1, label="line 1")
plt.plot(x, y2, label="line 2")

# plt.grid="true"

#plt.annotate("inp", xy=(3,9))

plt.legend()
# plt.legend(loc = "right")
plt.plot(x,y)

plt.show()
```
